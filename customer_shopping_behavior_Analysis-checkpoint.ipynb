```python
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
df["Review Rating"] = df.groupby("Category")["Review Rating"]\
                        .transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create column age_group 
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 25, 40, 60, 100],
    labels=["Young Adult", "Adult", "Middle Age", "Senior"]
)
```

```python
df.head(10)
```

```python
#create column purchase_frequency_days
frequency_mapping={"Fortnightly":14,"Weekly":7,"Monthly":30,"Bi-Weekly":14,"Annually":365,"Every 3 Months":90}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
                   
```

```python
df["purchase_frequency_days"],["frequency_of_purchases"]
```

```python
df=df.drop("promo_code_used",axis=1)
```
